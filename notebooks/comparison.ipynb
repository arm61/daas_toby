```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(palette='colorblind')
```

```python
labels = {
    '24 keV':4, 
    '12.5 keV':22,
    }
```

```python
plt.figure(figsize=(16, 10))
for i in labels.keys():
    ml = labels[i]
    file = h5py.File('output_analysis/ml_{}_chain_r.h5'.format(ml), 'r')
    chn = np.array(file['flatchain'])
    file.close()
    tail_thik = chn[:, [0] + list(range(4, chn.shape[1], 3))]
    head_solv = chn[:, [1] + list(range(5, chn.shape[1], 3))]
    tail_solv = chn[:, [2] + list(range(6, chn.shape[1], 3))]
    length = head_solv.shape[1]
    dh = np.tile(10, (length, 1)).T
    n_heads = (1 - head_solv) * dh / 319
    dt = tail_thik
    n_tails = (1 - tail_solv) * dt / 829
    plt.errorbar(range(length), (n_tails / n_heads).mean(axis=0),#/(n_tails / n_heads).mean(axis=0)[0], 
             (n_tails / n_heads).std(axis=0), marker='o', ls='--', label=i)
plt.ylabel('Tails/Heads')
plt.xlabel('Measurement')
plt.legend()
plt.show()
```
